```python
import cv2
```

```python
### IMPORTANT TO GET THE XML FILE

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')
```

```python
# IMAGE

image=cv2.imread('picture1.jpg')
```

```python
from matplotlib import pyplot as plt
plt.imshow(image)
```

```python
rgb_img = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
```

###### READ IMG CONVERT TO GRAYSCALE

```python
gr_img = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(gr_img)
```

##### DETECT COORDINATES OF FACES

```python
face=face_cascade.detectMultiScale(
    gr_img,
    scaleFactor=1.1,
    minNeighbors=5,
    minSize=(30,30)
)

print(face)
```

###### FACE DETECTION

```python
Image=cv2.imread('picture1.jpg')
for(x,y,w,h) in face:
    cv2.rectangle(Image,(x,y),(x+w,y+h),(255,0,0),20)
RGB_img=cv2.cvtColor(Image,cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
```

###### EXTRA CROP


```python
ims=[]
for (x,y,w,h) in face:
    crop=RGB_img[y:y+h,x:x+w]
    ims.append(crop)
plt.imshow(ims[0])
```
